```python
import numpy as np
import pandas as pd
```

```python
sen = pd.read_csv("Assignment5.csv",names= ['junior','senior'])
```

```python
sen
```

```python
mydata = dict(zip(sen.junior,sen.senior))
```

```python
result =[]
for name in mydata :
    temp = []
    temp.append(name)
    seni= mydata[name]
    if seni == mydata[seni]:
        if seni not in temp:
            temp.append(seni)
    while seni !=mydata[seni] :
        temp.append(seni)
        seni = mydata[seni]
        if seni == mydata[seni]:
            temp.append(seni)
    result.append(temp)

df = pd.DataFrame(result)
df
```
